```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from sklearn.manifold import LocallyLinearEmbedding


```

#Initial study of dataset

```python
X,y = load_digits(return_X_y=True)
```

```python
y.shape
```

#Resizing images to a square matrix

```python
plt.figure(figsize=(8,8))
plt.imshow(X[1].reshape(8,8),cmap='Greys')
```

#2D

```python
lle_model = LocallyLinearEmbedding(n_components=2, n_neighbors=15)
X_lle = lle_model.fit_transform(X, y=y)
data = pd.DataFrame({'LLE1':X_lle[:,0],'LLE2':X_lle[:,1],'class':y})
fig = plt.figure(figsize=(8,8))
sns.scatterplot(x='LLE1',y='LLE2',hue='class',data=data, palette='tab10')
```

#3D

```python
lle_model = LocallyLinearEmbedding(n_components=3, n_neighbors=15)
X_lle = lle_model.fit_transform(X, y=y)
data = pd.DataFrame({'LLE1':X_lle[:,0],'LLE2':X_lle[:,1],'LLE3':X_lle[:,2],'class':y})
```

```python
data.head()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection = '3d')
fig = px.scatter_3d(data, x='LLE1', y='LLE2', z='LLE3', color='class')
fig.show()
```
