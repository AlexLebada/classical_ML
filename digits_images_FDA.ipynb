```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

```

#Initial study of dataset

```python
X,y = load_digits(return_X_y=True)
```

```python
y.shape
```

#Resizing images to a square matrix

```python
plt.figure(figsize=(8,8))
plt.imshow(X[1].reshape(8,8),cmap='Greys')
```

```python
fig, axes= plt.subplots(nrows=2, ncols=3, sharex=True, sharey=True, figsize=(12,8))

axes[0,0].imshow(X[0].reshape(8,8),cmap='Greys')
axes[0,1].imshow(X[1].reshape(8,8),cmap='Greys')
axes[0,2].imshow(X[2].reshape(8,8),cmap='Greys')
axes[1,0].imshow(X[3].reshape(8,8),cmap='Greys')
axes[1,1].imshow(X[4].reshape(8,8),cmap='Greys')
axes[1,2].imshow(X[5].reshape(8,8),cmap='Greys')
plt.tight_layout()
```

#2D

```python
lda_model = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda_model.fit_transform(X, y=y)
data = pd.DataFrame({'FDA1':X_lda[:,0],'FDA2':X_lda[:,1],'class':y})
fig = plt.figure(figsize=(8,8))
sns.scatterplot(x='FDA1',y='FDA2',hue='class',data=data, palette='tab10')
fig.savefig('to.png')
```

#3D

```python
lda_model = LinearDiscriminantAnalysis(n_components=3)
X_lda = lda_model.fit_transform(X, y=y)
data = pd.DataFrame({'FDA1':X_lda[:,0],'FDA2':X_lda[:,1],'FDA3':X_lda[:,2],'class':y})
```

```python
data.head()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection = '3d')
fig = px.scatter_3d(data, x='FDA1', y='FDA2', z='FDA3', color='class')
fig.show()
```
